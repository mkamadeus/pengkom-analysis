```python
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir
```

```python
df = pd.DataFrame()
for file in listdir('./data'):
    df = df.append(pd.read_csv('./data/'+file), sort=True)
```

```python
df.head()
```

```python
df.columns
```

```python
for index, col in enumerate(df.columns):
    print('=====',index, df[col].unique())
```

```python
# df = df[['Course']]
df.sort_values(['Full name','Response'], inplace=True)
df.drop_duplicates(subset=['Full name'], inplace=True)
df
```

```python
selected_columns = ['Username', 'Q04', 'Q04_Materi Paling Sulit', 'Q05', 'Q05_Materi Paling Sulit','Q10',
       'Q10_Pernah Koding', 'Q11', 'Q11_Materi sesuai', 'Q12',
       'Q12_Materi sesuai']
df = df[selected_columns]
```

```python
df['Materi Termudah'] = df['Q04'].fillna(df['Q04_Materi Paling Sulit'])
df['Materi Tersulit'] = df['Q05'].fillna(df['Q05_Materi Paling Sulit'])
df['Pernah Ngoding'] = df['Q10'].fillna(df['Q10_Pernah Koding'])
df['Ngoding Berguna'] = df['Q11'].fillna(df['Q11_Materi sesuai'])
df['Materi Sesuai Dosen'] = df['Q12'].fillna(df['Q12_Materi sesuai'])
df
```

```python
df = df[['Username', 'Materi Termudah', 'Materi Tersulit', 'Pernah Ngoding', 'Ngoding Berguna', 'Materi Sesuai Dosen']]
df
```

```python
print(df['Materi Termudah'].unique())
print()
print(df['Materi Tersulit'].unique())
print()
print(df['Pernah Ngoding'].unique())
print()
print(df['Ngoding Berguna'].unique())
print()
print(df['Materi Sesuai Dosen'].unique())

df_nona = df.dropna()
df_nona
```

```python
df['Materi Termudah'].unique()
```

```python
mapping = {
    '1 : 1  - Input Output' : '1 Input Ouput',
    '1 : 1 - Input, Output, dan Percabangan': '1 Input Ouput',
    '2 : 2 - Larik dan Perulangan': '2 Pengulangan',
    '2 : 2 - Perulangan': '2 Pengulangan',
    '3 : 3 - Array dan Matriks': '3 Array dan Matriks',
    '3 : 3 - Fungsi dan Prosedur': '3 Array dan Matriks',
    '4 : 4 - Pengantar Analasis Data': '4 Prosedur dan Fungsi',
    '4 : 4 - Prosedur dan Fungsi': '4 Prosedur dan Fungsi',
    '5 : 5 - Analisis dan Visualisasi Data': '5 Analisis dan Visualisasi Data',
    '5 : 5 - Visualisasi Data': '5 Analisis dan Visualisasi Data'
}

df['Materi Termudah'] = df['Materi Termudah'].map(mapping)
df['Materi Tersulit'] = df['Materi Tersulit'].map(mapping)
df
```

```python
mapping = {
    0: 'Tidak',
    1: 'Ya',
    '2 : Tidak':'Tidak',
    '1 : Ya': 'Ya'
}

df_nona['Pernah Ngoding'] = df_nona['Pernah Ngoding'].map(mapping)
df_nona['Ngoding Berguna'] = df_nona['Ngoding Berguna'].map(mapping)
df_nona['Materi Sesuai Dosen'] = df_nona['Materi Sesuai Dosen'].map(mapping)
df_nona
```

```python
easy = df.groupby(['Materi Termudah']).count()['Username']
hard = df.groupby(['Materi Tersulit']).count()['Username']
```

```python
easy.plot(kind='bar', x='Materi Termudah', y='Username')
```

```python
fig, ax = plt.subplots(figsize=(15,8))
ax = hard.plot(kind='barh', x='Materi Tersulit', y='Username')
plt.ylabel('')
plt.title('Materi Tersulit Menurut Mahasiswa ITB 2019', fontsize=24)

for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width()/len(df['Username']))
    x = p.get_x() + p.get_width() +2
    y = p.get_y() + p.get_height()/2 -0.05
    ax.annotate(percentage, (x, y))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = (df_nona.groupby(['Pernah Ngoding']).count()['Username']).plot(kind='barh', x='Pernah Ngoding', y='Username')
plt.ylabel('')
plt.title('Apakah pernah ngoding?', fontsize=20)

for p in ax.patches:
    print('{:.1f}%'.format(100 * p.get_width()/len(df_nona['Username'])))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = (df_nona.groupby(['Materi Sesuai Dosen']).count()['Username']).plot(kind='barh', x='Materi Sesuai Dosen', y='Username')
plt.ylabel('')
plt.title('Apakah materi sesuai?', fontsize=20)

for p in ax.patches:
    print('{:.1f}%'.format(100 * p.get_width()/len(df_nona['Username'])))

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = (df_nona.groupby(['Ngoding Berguna']).count()['Username']).plot(kind='barh', x='Ngoding Berguna', y='Username')
plt.ylabel('')
plt.title('Ngoding berguna?', fontsize=20)

for p in ax.patches:
    print('{:.1f}%'.format(100 * p.get_width()/len(df_nona['Username'])))

plt.show()
```
